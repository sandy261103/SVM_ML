# Smote+TOMEK Method

This notebook will apply the SMOTE+TOMEK balancing method to the training data.

```python
import pandas as pd

df_train = pd.read_csv('train.csv')
X_train = df_train.drop('Diabetes_binary', axis=1)
y_train = df_train['Diabetes_binary']

print("Pre-sampling class imbalance:")
print(y_train.value_counts(normalize=True))
```

```python
from imblearn.combine import SMOTETomek

smote_tomek = SMOTETomek(random_state=92)
X_smotee_tomek, y_smote_tomek = smote_tomek.fit_resample(X_train, y_train)

print("Class distribution after SMOTE+TOMEK:")
print(y_smote_tomek.value_counts(normalize=True))
```
